```python
digit=pd.read_csv('train.csv')
```

```python
test=pd.read_csv('test.csv')
test
```

```python
from BasicNetwork import BasicNetwork3
```

```python
X=digit.drop(columns='label')
y=digit['label']
```

```python
clf=BasicNetwork3([784,100,10])
```

```python
training=X.iloc[:,:].values
labels=y.iloc[:].values
```

```python
clf.fit(training,labels,epochs=240,mini_batch_size=20)
```

```python
y_pred=clf.predict(training)
```

```python
from BasicNetwork import accuracy
accuracy(labels,y_pred)
```
